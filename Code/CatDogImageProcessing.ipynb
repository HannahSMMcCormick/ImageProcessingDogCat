```python
import pandas as pd
import numpy as np


from glob import glob

import cv2
import matplotlib.pylab as plt
```

# Reading Images

```python
dog_files = glob('../input/cat and dog/training_set/dogs/*.jpg')
cat_files = glob('../input/cat and dog/training_set/cats/*.jpg')



print(len(cat_files))
```

```python
img_mpl = plt.imread(cat_files[20])
img_cv2 = cv2.imread(cat_files[20])
img_mpl.shape, img_cv2.shape
```

```python
pd.Series(img_mpl.flatten()).plot(kind = 'hist',
                                  bins = 50,
                                  title = 'Distribution of Pixel Values')


```

# Display Image

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('off')
ax.imshow(img_mpl)
plt.show()
```

# Image Channel

```python
img_mpl.shape
```

```python
fig, axs = plt.subplots(1, 3, figsize = (15,5))
axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')

axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')

axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')

plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,5))

axs[0].imshow(img_cv2)
axs[1].imshow(img_mpl)

axs[0].axis('off')
axs[1].axis('off')

axs[0].set_title('CV2 Image')
axs[1].set_title('Matplotlib Image')

plt.show()
```

```python
img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots()
ax.imshow(img_cv2_rgb)
ax.axis('off')
plt.show()
```

# Image Manipulation 

```python
img = plt.imread(dog_files[4])
fig, ax = plt.subplots(figsize=(8,8))

ax.imshow(img)
ax.axis('off')
plt.show
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_gray.shape

fig,ax = plt.subplots(figsize=(8,8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('grey Image')
plt.show()
```
